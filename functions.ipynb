```python
#this is positional argument

def print_number(*args):
    for i in args:
        print(i)
```

```python
l=print_number(1,2,3,4,5,6,print_number(7,8,9))
print(l)
```

```python
#this is keyword argument 
def print_details(**kwargs):
    for keys,values in kwargs.items():
        print(f"{keys} {values}")
```

```python
print_details(name="hi",age="65")
```
