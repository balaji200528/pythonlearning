#packing and unpacking of tuple


```python
#packing
t=1,2,"b","c"
print(t)
```

```python
#unpacking
t=1,2,3,7,"b","k","k",2,3,4
a,b,c,d,*e,f=t #so whenever we use the same variable it will e overwrites the previous c 
print(a)
print(b)
print(c)
print(d)
print(e)
print(f)
```

NESTED TUPLES

```python
nested_tuple=((1,2,3),("a","b","c"),("r","f","i"))
print(nested_tuple[2][0:3])
```

In Python slicing, the syntax is:

list[start : stop : step]


start → inclusive (it starts at this index).

stop → exclusive (it stops before this index).

step → how much to jump (default = 1).

```python
l=(1,2,3,4,5,6)
print(l[0:4])
```

```python
l=[1,2,3,4,5]
print(l[0:4])
```

```python
nested_tuple=((1,2,3),("a","b","c"),("r","f","i"))

for sub in nested_tuple:
    for i in sub:
        print(i,end=" ")
    print()
```
